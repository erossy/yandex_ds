```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import ncf

x = np.linspace(0, 3, 300)
r = ncf.rvs(30,30,1,size=1000) # выборка длины n=1000 Нецентрального распределения Фишера (ncf) с параметрами (30,30,1)
plt.hist(r,30, density=1,label='Hist') # строим гистограмму распределения ncf
plt.plot(x, ncf(30,30,1).pdf(x),lw=4,label='Theor pdf') # строим график плотности распределения ncf
plt.legend(loc='best', frameon=True)
plt.title('NCF  PDF')
plt.ylabel('ncf_pdf,   w_i')# ncf_pdf - плотность ncf, w_i - относительная частота элементов выборки из нашего распределения ncf
plt.xlabel('x, xi')
```

Описание Нецентрального распределения Фишера (ncf):https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.ncf.html#scipy.stats.ncf
Функция распределения рассчитывается по формуле: ncf.pdf(x, df1, df2, nc) = exp(nc/2 + nc*df1*x/(2*(df1*x+df2))), где параметры df1, df2, nc > 0

```python
mean, var = ncf.mean(30,30,1), ncf.var(30,30,1)
mean, var  # воспользуемся функциями mean и var для расчета среднего и дисперсии исходного распределения 
        
```

```python
def sample_means(n): # создадим функцию генерации 1000 средних по выборкам из n элементов из нашего распределения ncf 
    sample_means = np.zeros(1000) # нет необходимости сохранять сами выборки 1000 по n, тк нам нужны только их средние
    for index in range(1000):
        sample_means[index] = ncf.rvs(30,30,1, size=n).mean()
    return sample_means
    
M3 = sample_means(3) # построим гистограмму по массиву M3 из 1000 средних из n=3 элементов
plt.hist(M3,30,density=1,label='Hist_n_3')
x = np.linspace(0.4,1.8,100)
plt.plot(x,sts.norm(1.1071, (0.1764/3)**0.5).pdf(x),lw=4,label='norm_theoretical')# и функцию плотности норм. распр. с mu = mean, sigma = (var/n)**0.5
plt.legend(loc='best', frameon=True)
plt.title('n=3')
plt.xlabel('x, x_i')
plt.ylabel('f(x),  w_i') # f(x) - плотность норм, w_i - относительная частота средних по выборкам из n элементов из нашего распределения ncf
```

```python
M10 = sample_means(10) # построим гистограмму по массиву M10 из 1000 средних из n=10 элементов
plt.hist(M10,30,density=1,label='Hist_n_10')
x = np.linspace(0.5,1.7,100)
plt.plot(x,sts.norm(1.1071, (0.1764/10)**0.5).pdf(x),lw=4,label='norm_theoretical')# и функцию плотности норм. распр. с mu = mean, sigma = (var/n)**0.5
plt.legend(loc='best', frameon=True)
plt.title('n=10')
plt.xlabel('x,  x_i')
plt.ylabel('f(x),  w_i') # f(x) - плотность норм, w_i - относительная частота средних по выборкам из n элементов из нашего распределения ncf
```

```python
M80 = sample_means(80) # построим гистограмму по массиву M80 из 1000 средних из n=80 элементов
plt.hist(M80,30,density=1,label='Hist_n_80')
x = np.linspace(0.9,1.3,100)
plt.plot(x,sts.norm(1.1071, (0.1764/80)**0.5).pdf(x),lw=4,label='norm_theoretical')# и функцию плотности норм. распр. с mu = mean, sigma = (var/n)**0.5
plt.legend(loc='best', frameon=True)
plt.title('n=80')
plt.xlabel('x, x_i')
plt.ylabel('f(x),  w_i') # f(x) - плотность норм, w_i - относительная частота средних по выборкам из n элементов из нашего распределения ncf
```

ВЫВОД: 
Как видим, при n = 3 еще наблюдается ассиметрия, гистограмма напоминает плотность изначального распределения. При увеличении n до n=80  - гистограмма распределения выборочных средних отлично апроксимируется нормальным распределением. Таким образом, точность аппроксимации распределения выборочных средних нормальным распределением увеличивается с ростом n, что наглядно демонстрирует Центральную предельную теорему.
